```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, roc_auc_score

def load_data():
    # Correct file path based on user's input
    file_path = '/content/drive/MyDrive/MLproject/WA_Fn-UseC_-HR-Employee-Attrition.csv'

    # Check if the file exists
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File not found at: {file_path}. Please check the path and ensure the file exists.")

    # Read the dataset
    data = pd.read_csv(file_path)

    return data

def preprocess_data(data):
    # Handle missing values
    data = data.dropna()

    # Encode categorical variables (you can use OneHotEncoder, LabelEncoder, etc.)
    data = pd.get_dummies(data)

    # Split data into features and target variable
    X = data.drop('Attrition_Yes', axis=1)  # Assuming 'Attrition_Yes' is the target
    y = data['Attrition_Yes']  # Target column

    # Split into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Standardize features
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    return X_train, X_test, y_train, y_test

def train_models(X_train, X_test, y_train, y_test):
    # Logistic Regression
    logistic_regression = LogisticRegression(max_iter=1000)
    logistic_regression.fit(X_train, y_train)
    y_pred_lr = logistic_regression.predict(X_test)
    auc_lr = roc_auc_score(y_test, y_pred_lr)
    print(f"Logistic Regression AUC Score: {auc_lr:.4f}")
    print(classification_report(y_test, y_pred_lr))

    # Decision Tree
    decision_tree = DecisionTreeClassifier()
    decision_tree.fit(X_train, y_train)
    y_pred_dt = decision_tree.predict(X_test)
    auc_dt = roc_auc_score(y_test, y_pred_dt)
    print(f"Decision Tree AUC Score: {auc_dt:.4f}")
    print(classification_report(y_test, y_pred_dt))

    # Support Vector Machine
    svm = SVC(probability=True)
    svm.fit(X_train, y_train)
    y_pred_svm = svm.predict(X_test)
    auc_svm = roc_auc_score(y_test, y_pred_svm)
    print(f"SVM AUC Score: {auc_svm:.4f}")
    print(classification_report(y_test, y_pred_svm))

def main():
    # Load the dataset
    data = load_data()

    # Preprocess the data
    X_train, X_test, y_train, y_test = preprocess_data(data)

    # Train and evaluate models
    train_models(X_train, X_test, y_train, y_test)

if __name__ == '__main__':
    main()
```
